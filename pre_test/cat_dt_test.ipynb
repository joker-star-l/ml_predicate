```python
## 使用Sklearn调用GBDT模型拟合数据并可视化
import numpy as np
import pydotplus
from sklearn.tree import DecisionTreeRegressor

X = np.eye(10, dtype=np.float32)
X2 = np.random.randn(10, 2)
X = np.concatenate((X, X2), axis=1)
print(X)
y = np.array([5.16, 4.73, 5.95, 6.42, 6.88, 7.15, 8.95, 8.71, 9.50, 9.15])
# y = np.random.randn(100)
print(y)
gbdt = DecisionTreeRegressor(max_depth=10).fit(X, y)
```

```python
from IPython.display import Image  
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz

def plot_tree():
    dot_data = export_graphviz(gbdt, out_file=None, filled=True, rounded=True, special_characters=True, precision=2)
    return graph_from_dot_data(dot_data)  
```

```python
# mean + learning_rate * leaf_value
gbdt.predict(X)
```

```python
Image(plot_tree().create_png())
```
