```python
file = 'branch_test2.csv'
leafs = 2 ** 10
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv(file)
df
```

```python
data = df[df['leafs'] == leafs]['cost'].groupby(df['prediction'], sort=True).mean().sort_index()
data.index
```

```python
end = data.shape[0]

# x = data.index[:end + 1]
x = np.arange(0, end)
y = data.values[:end + 1]
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, s=2)
plt.ylabel('time (s)')
plt.grid()
plt.ylim(0)
```
