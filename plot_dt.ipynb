```python
model_name = 'house_16H_d5_l24_n47_20241108023539'
joblib_path = f'model/{model_name}.joblib'
pruned_joblib_path = f'model_output/{model_name}.joblib'
```

```python
import joblib
import numpy as np
from IPython.display import Image  
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz

model = joblib.load(joblib_path)
dot_data = export_graphviz(model, out_file=None, filled=True, rounded=True, special_characters=True, precision=3)
graph = graph_from_dot_data(dot_data)
# print(model.predict_proba(np.array([24., 0., 0., 109., 0., 9., 0.]).reshape(1, -1)))
Image(graph.create_png())
```
